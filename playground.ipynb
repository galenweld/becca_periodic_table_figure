```python
# keep just the elements that are in XRF, and drop the others
# then normalize the non-XRF columns so they sum to 1
print(data.dropna(subset=('XRF',), axis='index').sum())
data = data.dropna(subset=('XRF',), axis='index') / data.dropna(subset=('XRF',), axis='index').sum()
print(data.dropna(subset=('XRF',), axis='index').sum())
print(data)
```

```python
data = data.fillna(0.0)
```

```python
data2 = data2.dropna(subset=('XRF',), axis='index') / data.dropna(subset=('XRF',), axis='index').sum()
data2 = data2.fillna(0.0)

# normalize by column sum
data2 = data2 / data2.sum(axis='rows')

data2
```

```python
def plot_key(ax, symbol, grayness=.5):

    sns.heatmap([2*[grayness,],2*[grayness,]],
        cmap  = 'Greys',
        cbar  = False,
        vmin  = 0, #data.min().min(),
        vmax  = 1, #data.max().max(),
        annot = np.array([['XRF', 'XPS'], ['ICP', 'EDS']]),
        fmt   = '',
        ax    = ax,
        annot_kws = {"fontsize": "large"},
    )

    # label the axes with the symbol
    ax.text(
        x         = .5,
        y         = .5,
        s         = symbol,
        transform = ax.transAxes,
        ha        = 'center',
        va        = 'center',
        fontsize  = 22
    )
    
    # turn off ticks
    ax.set_xticks([])
    ax.set_yticks([])
    
    # make frame visible
    for _, spine in ax.spines.items():
        spine.set_visible(True)
```

```python
fig, ax = plt.subplots(figsize=(2,2))

plot_key(ax, 'M', grayness=.25)

plt.savefig('figs/legend.pdf')
plt.show()
```

```python
def plot_element(ax, symbol, XRF=np.nan, XPS=np.nan, ICP=np.nan, EDS=np.nan):
    
    if all(map(np.isnan, [XRF, XPS, ICP, EDS])):
        # if we don't have any data, set background to grey
        ax.set_facecolor('lightgrey')
    
    else:
        sns.heatmap(
            [[XRF, XPS],
             [ICP, EDS]],
            cmap  = 'OrRd',
            cbar  = False,
            vmin  = 0, #data.min().min(),
            vmax  = 1, #data.max().max(),
            annot = True,
            fmt   = '.2f',
            ax    = ax,
            annot_kws = {"fontsize": "large"},
        )
    
    # label the axes with the symbol
    ax.text(
        x         = .5,
        y         = .5,
        s         = symbol,
        transform = ax.transAxes,
        ha        = 'center',
        va        = 'center',
        fontsize  = 22
    )
    
    # turn off ticks
    ax.set_xticks([])
    ax.set_yticks([])
    
    # make frame visible
    for _, spine in ax.spines.items():
        spine.set_visible(True)
```

```python
fig, ax = plt.subplots(figsize=(2,2))

plot_element(ax, 'Mn', 0.4, .5, .6, .7)

plt.show()
```

```python
fig, axs = plt.subplots(nrows=1, ncols=5, figsize=(10,2))

for ax, symbol in zip(axs.flatten(order='C'), df.iloc[3:4,6:11].to_numpy().flatten(order='c')):    
    ax.set_aspect('equal', 'box')
    try:
        plot_element(ax, symbol, **data.loc[symbol,:].to_dict())
    except KeyError:
        plot_element(ax, symbol)
        
plt.savefig('figs/cobalt-rich-black-mass-sample.pdf')
plt.show()
```

```python
fig, axs = plt.subplots(nrows=2, ncols=6, figsize=(12,4))

# top row
for ax, symbol in zip(axs[0,1:], df.iloc[3:4,6:11].to_numpy().flatten(order='c')):   
    ax.set_aspect('equal', 'box')
    try:
        plot_element(ax, symbol, **data.loc[symbol,:].to_dict())
    except KeyError:
        plot_element(ax, symbol)

# bottom row
for ax, symbol in zip(axs[1,1:], df.iloc[3:4,6:11].to_numpy().flatten(order='c')):   
    ax.set_aspect('equal', 'box')
    try:
        plot_element(ax, symbol, **data2.loc[symbol,:].to_dict())
    except KeyError:
        plot_element(ax, symbol)
        
# legend
axs[0,0].set_aspect('equal', 'box')
plot_key(axs[0,0], 'M', grayness=.25)

# hide lower left
axs[1,0].axis('off')


# increase space between rows
fig.subplots_adjust(hspace=0.4)


# add labels
# figure coords: (0,0) is the bottom left and (1,1) is the top right of the figure.
fig.text(0.13, .95, 'mass %', fontsize=16, ha='left', va='top')
fig.text(0.26, .95, 'Co-rich black mass sample', fontsize=16, ha='left', va='top')
fig.text(0.26, .51, 'Mn-rich black mass sample', fontsize=16, ha='left', va='top')

plt.savefig('figs/combined.pdf')
plt.show()
```
